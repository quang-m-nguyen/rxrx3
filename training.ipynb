```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt
import umap

# Check for MPS (Apple Silicon) device
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

# Custom Dataset for Cell Images
class CellImageDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.image_paths = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))]
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")
        if self.transform:
            aug1 = self.transform(img)  # First augmentation
            aug2 = self.transform(img)  # Second augmentation
        return aug1, aug2

# Define Augmentations
transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),  # Mild crop to preserve cell structure
    transforms.RandomRotation(90),  # Cells often lack fixed orientation
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Load Dataset
image_dir = "/Volumes/External SSD/rxrx3_resized/rxrx3_512/mixed_images/"  # Replace with your folder path
dataset = CellImageDataset(image_dir, transform=transform)
data_loader = DataLoader(dataset, batch_size=64, shuffle=True, num_workers=0, pin_memory=True)  # Set num_workers=0

# Define SimCLR Model
class SimCLR(nn.Module):
    def __init__(self, base_model, out_dim=128):
        super(SimCLR, self).__init__()
        self.encoder = base_model
        self.encoder.fc = nn.Identity()  # Remove classification head
        # Projection head: 512 -> 512 -> 128
        self.projection_head = nn.Sequential(
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, out_dim)
        )

    def forward(self, x):
        h = self.encoder(x)  # Feature vector (512-dim)
        z = self.projection_head(h)  # Projected vector (128-dim)
        return h, z

# NT-Xent Loss (Simplified)
def nt_xent_loss(z1, z2, temperature=0.5):
    batch_size = z1.size(0)
    z = torch.cat([z1, z2], dim=0)  # [2N, 128]
    sim_matrix = torch.exp(torch.nn.functional.cosine_similarity(z.unsqueeze(1), z.unsqueeze(0), dim=2) / temperature)
    mask = torch.eye(2 * batch_size, dtype=torch.bool, device=device)
    sim_matrix = sim_matrix.masked_fill(mask, 0)  # Remove self-similarity
    pos_sim = torch.exp(torch.nn.functional.cosine_similarity(z1, z2) / temperature)
    loss = -torch.log(pos_sim / (sim_matrix.sum(dim=1)[:batch_size] + 1e-6)).mean()
    return loss

# Initialize Model with updated weights argument
base_model = models.resnet18(weights="ResNet18_Weights.IMAGENET1K_V1")  # Updated syntax
model = SimCLR(base_model).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.0003)  # Small LR for fine-tuning

# Training Loop with Visualization
num_epochs = 10  # Start small, increase later
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for batch_idx, (aug1, aug2) in enumerate(data_loader):
        aug1, aug2 = aug1.to(device), aug2.to(device)
        optimizer.zero_grad()
        h1, z1 = model(aug1)  # Features and projections
        h2, z2 = model(aug2)
        loss = nt_xent_loss(z1, z2)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        
        # Print progress
        if batch_idx % 10 == 0:
            print(f"Epoch [{epoch}/{num_epochs}], Batch [{batch_idx}/{len(data_loader)}], Loss: {loss.item():.4f}")
    
    avg_loss = total_loss / len(data_loader)
    print(f"Epoch [{epoch}/{num_epochs}] Average Loss: {avg_loss:.4f}")

    # Visualization after every 2 epochs
    if epoch % 2 == 0:
        model.eval()
        features = []
        with torch.no_grad():
            for aug1, _ in DataLoader(dataset, batch_size=64, shuffle=False, num_workers=0):
                aug1 = aug1.to(device)
                h, _ = model(aug1)  # Extract features (512-dim)
                features.append(h.cpu().numpy())
        features = np.concatenate(features, axis=0)
        
        # Reduce to 2D with UMAP
        reducer = umap.UMAP(n_components=2, random_state=42)
        embedding = reducer.fit_transform(features)
        
        # Plot
        plt.figure(figsize=(8, 6))
        plt.scatter(embedding[:, 0], embedding[:, 1], s=5, alpha=0.5)
        plt.title(f"UMAP of Features - Epoch {epoch}")
        plt.xlabel("UMAP 1")
        plt.ylabel("UMAP 2")
        plt.show()
        model.train()

# Save the trained encoder
torch.save(model.encoder.state_dict(), "resnet18_encoder.pth")
print("Training complete! Encoder saved.")
```
